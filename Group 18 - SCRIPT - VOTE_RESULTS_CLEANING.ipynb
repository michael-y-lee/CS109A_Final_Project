```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("data/1976-2018-house.csv", encoding = 'unicode_escape', thousands=',')
```

```python
data_2000_to_2018 = data[data["year"]>=2000]
data_2000_to_2018 = data_2000_to_2018[data_2000_to_2018["year"]<=2018]
```

```python
data_2000_to_2018
```

```python
data_2000_to_2018['democrat_vote'] = np.where(data_2000_to_2018['party'] == "democrat", data_2000_to_2018['candidatevotes'], 0.)
data_2000_to_2018['republican_vote'] = np.where(data_2000_to_2018['party'] == "republican", data_2000_to_2018['candidatevotes'], 0.)

data_2000_to_2018_cleaned = data_2000_to_2018[["year","state", "state_po","district","democrat_vote","republican_vote","totalvotes"]]

data_2000_to_2018_cleaned['democrat_vote']=data_2000_to_2018_cleaned['democrat_vote'].values.astype('int64')
```

```python
data_2000_to_2018_cleaned
```

# Group data by Year, State and District

```python
cleaned_df = data_2000_to_2018_cleaned.groupby(['year','state_po','district']).agg('max')
```

```python
cleaned_df['other_vote']= cleaned_df['totalvotes']-cleaned_df['democrat_vote']-cleaned_df['republican_vote']

cleaned_df = cleaned_df.reindex(columns=['democrat_vote', 'republican_vote', 'other_vote', 'totalvotes'])
```

```python
cleaned_df.to_csv("cleaned_votes.csv")
```

# Election Winner

```python
cleaned_df['winner'] = 999

cleaned_df['percent_democrat'] = 999
cleaned_df['percent_republican'] = 999
cleaned_df['percent_independent'] = 999

for row in range(0,cleaned_df.shape[0],1):
    #dem
    check_val = cleaned_df.iloc[row,0]/cleaned_df.iloc[row,3]
    if check_val > 0:
        cleaned_df.iloc[row,5] = check_val
    else: 
        cleaned_df.iloc[row,5] = 0
        
for row in range(0,cleaned_df.shape[0],1):
    #rep
    check_val = cleaned_df.iloc[row,1]/cleaned_df.iloc[row,3]
    if check_val > 0:
        cleaned_df.iloc[row,6] = check_val
    else: 
        cleaned_df.iloc[row,6] = 0
        
for row in range(0,cleaned_df.shape[0],1):
    #ind
    check_val = cleaned_df.iloc[row,2]/cleaned_df.iloc[row,3]
    if check_val > 0:
        cleaned_df.iloc[row,7] = check_val
    else: 
        cleaned_df.iloc[row,7] = 0
        
for row in range(0,cleaned_df.shape[0],1):
    if (cleaned_df.iloc[row,5] > cleaned_df.iloc[row,6]) and (cleaned_df.iloc[row,5] > cleaned_df.iloc[row,7]):
        cleaned_df.iloc[row,4] = 0 #democrat
    elif (cleaned_df.iloc[row,6] > cleaned_df.iloc[row,5]) and (cleaned_df.iloc[row,6] > cleaned_df.iloc[row,7]):
        cleaned_df.iloc[row,4] = 1 #republican
    elif (cleaned_df.iloc[row,7] > cleaned_df.iloc[row,5]) and (cleaned_df.iloc[row,7] > cleaned_df.iloc[row,6]):
        cleaned_df.iloc[row,4] = 2 #independent
    else:
        cleaned_df.iloc[row,4] = 9999 #Its broken!

cleaned_df = cleaned_df.drop(['percent_democrat', 'percent_republican', 'percent_independent'], axis=1)
```

# Export and reset "groupby"

```python
# write and read file, to encode group by
cleaned_df.to_csv("cleaned_votes.csv")
cleaned_df = pd.read_csv("cleaned_votes.csv", encoding = 'unicode_escape', thousands=',')
```

# Midterm Years

```python
# Flag Midterm Years
cleaned_df['midterm'] = 999

midterm_years = [2018, 2014, 2010, 2006, 2002]

for row in range(0,cleaned_df.shape[0],1):
    if (cleaned_df.iloc[row,0] == 2018) or (cleaned_df.iloc[row,0] == 2014) or (cleaned_df.iloc[row,0] == 2010) or (cleaned_df.iloc[row,0] == 2006) or (cleaned_df.iloc[row,0] == 2002):
        cleaned_df.iloc[row,8] = 1
    else:
        cleaned_df.iloc[row,8] = 0
```

# Party in Power - Incumbent President's Party

```python
# Incumbent - President
cleaned_df['incumbent_president'] = 999

for row in range(0,cleaned_df.shape[0],1):
    if (cleaned_df.iloc[row,0] >= 2017):
        cleaned_df.iloc[row,9] = 1 #republican - Trump
    if (cleaned_df.iloc[row,0] <= 2016):
        cleaned_df.iloc[row,9] = 0 #democrat - Obama
    if (cleaned_df.iloc[row,0] <= 2008):
        cleaned_df.iloc[row,9] = 1 #republican - Bush
    if (cleaned_df.iloc[row,0] <= 2000):
        cleaned_df.iloc[row,9] = 0 #democrat - Clinton
```

# Mean Reversion in the Midterms - Turnovers

```python
cleaned_df['turn_over'] = 999
cleaned_df['district_turn_over'] = 999
cleaned_df['new_district'] = 999

for row in range(0,cleaned_df.shape[0],1): #run entire dataset
# for row in range(436,871,1): # try for just 2002  
    year = cleaned_df.iloc[row,0] #the current year
    state = cleaned_df.iloc[row,1] #the current state
    district = cleaned_df.iloc[row,2] #the current district
    current_winner = cleaned_df.iloc[row,7] # the current winner
    if (year != 2000): #there is no 1998, so only years above 2002
        temp_subset = cleaned_df.copy() #subset dataframe
        temp_subset = temp_subset[temp_subset["year"] == (year-2)] # find last elections year 
        temp_subset = temp_subset[temp_subset["state_po"] == state] #for same state
        temp_subset = temp_subset[temp_subset["district"] == district] #and same district - SHOULD YIELD ONLY ONE RESULT
        if temp_subset.empty: #if there is no result, due to redistricting before 2010 census
            cleaned_df.iloc[row,10] = 0 # there was no 'turn_over'
            cleaned_df.iloc[row,11] = 0 # there was no turnover 'district_turnover' 
            cleaned_df.iloc[row,12] = 1 # a new district was created
        else: # if there WAS a past election found
            cleaned_df.iloc[row,12] = 0 # is an existing district
            if temp_subset["winner"].all() == current_winner: # if district hasnt changed hands (dem stayed dem)
                cleaned_df.iloc[row,10] = 0 #turn_over flag, no
                cleaned_df.iloc[row,11] = 0 #district_turnover flag, no       
            elif temp_subset["winner"].all() != current_winner: # if district did change hands (dem became rep)
                cleaned_df.iloc[row,11] = 1 #district_turnover flag 
                if temp_subset["winner"].all() != temp_subset["incumbent_president"].all(): # if president party lost a seat
                    cleaned_df.iloc[row,10] = 1 #turn_over flag = yes, was swing away from party in power
                else: #presidents party did NOT lose a seat
                    cleaned_df.iloc[row,10] = 0 #turn_over flag, no
            else:
                cleaned_df.iloc[row,10] = 9999 #error flag for 'turn_over'
                cleaned_df.iloc[row,11] = 9999 #error flag for 'district_turnover' 
    else:
        cleaned_df.iloc[row,10] = 9999 #error flag for 'turn_over'
        cleaned_df.iloc[row,11] = 9999 #error flag for 'district_turnover'       
```

```python
# DELETE YEAR 2000 - RESULTS NO LONGER USEFUL
cleaned_df = cleaned_df[cleaned_df["year"] != 2000]
```
