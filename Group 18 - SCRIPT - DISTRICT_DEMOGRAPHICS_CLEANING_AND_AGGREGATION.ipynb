# DEMOGRAPHICS BY STATE AND DISTRICT - ETL

```python
import os
import glob

import numpy as np
import pandas as pd
```

```python
extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
all_filenames.remove("combined_district_demographics.csv") #don't include the combined file, creates redundant rows
```

```python
all_filenames
```

# Combined all CSV Files into one Dataframe

```python
#combine all files in the list
combined_csv = pd.concat([pd.read_csv(f).T for f in all_filenames]) #transpose existing dataframe - districts as observations
#export to csv
combined_csv.to_csv("combined_district_demographics.csv", index=True, encoding='utf-8-sig')
```
